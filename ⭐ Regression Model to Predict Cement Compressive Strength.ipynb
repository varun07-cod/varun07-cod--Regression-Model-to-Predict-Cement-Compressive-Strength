```python
# import library
import pandas as pd
```

```python
# import data
cement = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Concrete%20Compressive%20Strength.csv')
```

```python
# visualize pairplot

import seaborn as s
import matplotlib.pyplot as plt
s.pairplot(cement)
```

```python
# columns name
cement.columns
```

```python
# define y
y = cement['Concrete Compressive Strength(MPa, megapascals) ']
```

```python
# define X
x = cement[['Cement (kg in a m^3 mixture)',
       'Blast Furnace Slag (kg in a m^3 mixture)',
       'Fly Ash (kg in a m^3 mixture)',
       'Water (kg in a m^3 mixture)',
       'Superplasticizer (kg in a m^3 mixture)',
       'Coarse Aggregate (kg in a m^3 mixture)',
       'Fine Aggregate (kg in a m^3 mixture)',
       'Age (day)']]
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

```python
# verify shape
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train model
model.fit(x_train,y_train)
```

```python
# predict with model
y_pred=model.predict(x_test)
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
# model MAE
mean_absolute_error(y_test,y_pred)
```

```python
# model MAPE
mean_absolute_percentage_error(y_test,y_pred)
```

```python
# model MSE
mean_squared_error(y_test,y_pred)
```

```python
# future prediction
x_predict = list(model.predict(x_test))
predicted_df = {'predicted_values': x_predict, 'original_values': y_test}
pd.DataFrame(predicted_df).head(20)
```
